```python
def normalize_name(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip().lower()
    s = " ".join(s.split())
    s = unicodedata.normalize("NFD", s)
    s = "".join(ch for ch in s if unicodedata.category(ch) != "Mn")
    return s
```

```python
def extract_model_tables(model):
    coef = model.summary2().tables[1].reset_index().rename(columns={"index": "term"})
    fit = pd.DataFrame({
        "metric": ["n", "r_squared", "adj_r_squared", "f_statistic", "f_pvalue", "aic", "bic"],
        "value": [
            int(model.nobs),
            model.rsquared,
            model.rsquared_adj,
            model.fvalue,
            model.f_pvalue,
            model.aic,
            model.bic
        ]
    })
    return coef, fit
```

$A = \beta (\Delta KT_{post} + \Delta TD_{post} + \Delta HV_{post})+\varepsilon$

```python
matched["dKT"] = matched["KT_post"] - matched["KT_pre"]
matched["dTD"] = matched["TD_post"] - matched["TD_pre"]
matched["dHV"] = matched["HV_post"] - matched["HV_pre"]
matched["dA"]  = matched["A_post"]  - matched["A_pre"]
```

```python
model_gain = smf.ols("dA ~ dKT + dTD + dHV", data=matched).fit()

print(model_gain.summary())

gain_coef, gain_fit = extract_model_tables(model_gain)

print("Bảng hệ số — Cách 2")
display(gain_coef)

print("Độ phù hợp mô hình — Cách 2")
display(gain_fit)

```
